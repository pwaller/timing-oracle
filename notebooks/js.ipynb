```python
import numpy as N, scipy as S, matplotlib.pylab as P

def f(path):
    data = N.array([int(x) / 1e5 for x in open("../logs/js/{}.log".format(path)) if x.strip()])
    # discard first 150 samples
    data = data[150:]
    return data
```

```python
fig, axes = P.subplots(4, 5, figsize=(25, 7)) #, sharey=True)

plots = "x y ya z za zx zy zz zza zzy zzz zzza zzzzz zzzzzz zzzzzzz zzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzzz".split()

for a, p in zip(axes.flat, plots):
    a.set_title(p)
    data = f(p)[::10]
    a.hist(data, bins=300, range=(4, 12))
    a.axvline(data.mean(), color="r")
    
print len(data)
```
